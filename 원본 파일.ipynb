```python
try:
    nltk.data.find('tokenizers/punkt')
except nltk.downloader.DownloadError:
    print("NLTK 'punkt' 토크나이저 다운로드 중...")
    nltk.download('punkt')
except LookupError:
    print("NLTK 'punkt' 토크나이저 다운로드 중...")
    nltk.download('punkt')

try:
    nltk.data.find('taggers/averaged_perceptron_tagger')
except nltk.downloader.DownloadError:
    print("NLTK 'averaged_perceptron_tagger' 다운로드 중...")
    nltk.download('averaged_perceptron_tagger')
except LookupError:
     print("NLTK 'averaged_perceptron_tagger' 다운로드 중...")
     nltk.download('averaged_perceptron_tagger')


def get_pos_distribution(text):
    """
    텍스트를 입력받아 품사(POS) 태그 분포(비율)를 계산합니다.
    영어 텍스트를 기준으로 NLTK의 POS 태거를 사용합니다.
    """
    if not isinstance(text, str) or not text.strip(): #
        return {}


    cleaned_text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE) 
    cleaned_text = re.sub(r'<.*?>', '', cleaned_text) 
    cleaned_text = cleaned_text.lower()


    words = nltk.word_tokenize(cleaned_text)

    if not words: 
        return {}

    # POS

    pos_tags = nltk.pos_tag(words)

    # 품사 tag 빈도 계산
    tags = [tag for (word, tag) in pos_tags]
    tag_counts = Counter(tags)

    # 전체 단어 수 
    total_tags = len(tags) # 

    # 품사별 비율 계산
    pos_distribution = {tag: count / total_tags for tag, count in tag_counts.items()}

    return pos_distribution


data['Answer_safe'] = data['Answer'].fillna('')

pos_series = data['Answer_safe'].apply(get_pos_distribution)

pos_df = pd.DataFrame(pos_series.tolist())

data_with_pos = pd.concat([data, pos_df], axis=1).fillna(0)

data_with_pos = data_with_pos.drop(columns=['Answer_safe'])

print("--- 품사 분포 분석 결과 (샘플) ---")
print(data_with_pos.head())
print(f"\n총 {len(pos_df.columns)}개의 고유 품사 태그가 발견되었습니다.")


if 'model' in data_with_pos.columns:

    pos_tag_columns = pos_df.columns.tolist() 

    avg_pos_by_model = data_with_pos.groupby('model')[pos_tag_columns].mean()

    print("\n--- 모델별 평균 품사 비율 ---")
    # 중요한 몇 가지 품사 태그만 출력하거나, 상위 N개 태그를 선택하여 출력 가능
    # Penn Treebank Tag Set (자주 나오는 것들):
    # NN: 명사, 단수/질량형
    # NNS: 명사, 복수
    # NNP: 고유명사, 단수
    # NNPS: 고유명사, 복수
    # VB: 동사, 기본형
    # VBD: 동사, 과거형
    # VBG: 동사, 현재분사/동명사
    # VBN: 동사, 과거분사
    # VBP: 동사, 현재형 (1, 2인칭 단/복수; 3인칭 복수)
    # VBZ: 동사, 현재형 (3인칭 단수)
    # JJ: 형용사
    # JJR: 형용사, 비교급
    # JJS: 형용사, 최상급
    # RB: 부사
    # RBR: 부사, 비교급
    # RBS: 부사, 최상급
    # IN: 전치사 또는 종속접속사
    # DT: 한정사 (a, the, this, that 등)
    # PRPS: 소유격 대명사 (my, his, her, our, their)
    # PRP: 인칭 대명사 (I, he, she, it, we, they, you)
    # CC: 등위접속사 (and, but, or)
    # CD: 기수 (숫자)
    # WDT: wh-한정사 (which, what)
    # WP: wh-대명사 (who, what, whom, which)
    # WRB: wh-부사 (where, when, why)

    relevant_pos_tags = ['NN', 'NNS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ',
                         'JJ', 'RB', 'IN', 'DT', 'PRP', 'PRPS', 'CC']
    present_relevant_pos_tags = [tag for tag in relevant_pos_tags if tag in avg_pos_by_model.columns]

    if present_relevant_pos_tags:
         print(avg_pos_by_model[present_relevant_pos_tags].to_string(float_format='%.4f'))
         tags_to_plot = ['NN', 'VB', 'JJ', 'RB', 'PRP', 'IN'] 
         present_tags_to_plot = [tag for tag in tags_to_plot if tag in data_with_pos.columns]

         if present_tags_to_plot:
             n_tags_plot = len(present_tags_to_plot)
             n_cols_plot = 2
             n_rows_plot = (n_tags_plot + n_cols_plot - 1) // n_cols_plot

             fig_pos, axes_pos = plt.subplots(n_rows_plot, n_cols_plot, figsize=(7 * n_cols_plot, 5 * n_rows_plot))
             axes_pos = axes_pos.flatten() if n_tags_plot > 1 else [axes_pos] 

             unique_models = sorted(data_with_pos['model'].unique()) 
             palette = sns.color_palette("husl", len(unique_models)) 

             for i, tag in enumerate(present_tags_to_plot):
                 sns.boxplot(data=data_with_pos, x='model', y=tag, ax=axes_pos[i], palette=palette, showfliers=False, order=unique_models)
                 axes_pos[i].set_title(f'Distribution of "{tag}" (POS Tag)', fontsize=14)
                 axes_pos[i].set_xlabel('Model', fontsize=12)
                 axes_pos[i].set_ylabel(f'Proportion of "{tag}"', fontsize=12)
                 axes_pos[i].tick_params(axis='x', rotation=45, labelsize=10)
                 axes_pos[i].tick_params(axis='y', labelsize=10)

             # 빈 subplot 제거
             for j in range(i + 1, len(axes_pos)): fig_pos.delaxes(axes_pos[j])
             fig_pos.suptitle('Comparison of Part-of-Speech Tag Proportions by Model (Box Plots)', fontsize=18, y=1.03)
             plt.tight_layout(rect=[0, 0.03, 1, 0.99]) 
             plt.show()
         else:
             print("\n선택된 대표 품사 태그가 데이터에 존재하지 않습니다.")

    else:
         print("\n데이터에 품사 태그 컬럼이 존재하지 않습니다. get_pos_distribution 함수의 반환 값을 확인하세요.")

else:
    print("오류: 'data' DataFrame에 'model' 컬럼이 없습니다.")
```

```python
def count_unique_words(text):
    """텍스트 내 고유 단어 수 계산 (간단 전처리 포함)"""
    if not isinstance(text, str): return 0
    words = re.sub(r'[^a-z0-9\s]', '', str(text).lower()).split()
    return len(set(w for w in words if w)) 
```

```python
def preprocess_text(text):
    """텍스트 전처리: 소문자, 구두점 제거, 불용어 제거"""
    if not isinstance(text, str):
        return ""
    text = text.lower() 
    text = re.sub(r'[^a-z0-9\s]', '', text)
    words = text.split() # 토큰화 (공백 기준)
    # 불용어 제거 및 한 글자 단어 제거
    words = [word for word in words if word not in stop_words_en and len(word) > 1]
    return ' '.join(words)

data['processed_answer'] = data['Answer'].fillna('').apply(preprocess_text)
```

```python
vectorizer = TfidfVectorizer(min_df=2, max_df=0.90, ngram_range=(1, 2)) # 유니그램 및 바이그램 포함
tfidf_matrix = vectorizer.fit_transform(data['processed_answer'])
feature_names = vectorizer.get_feature_names_out() # 단어(피처) 목록


tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names, index=data.index)
```

```python
import pandas as pd



if 'model' in data.columns:
    tfidf_renamed_df = tfidf_df.copy() 
    if 'model' in tfidf_renamed_df.columns:
        print("경고: TF-IDF DataFrame에 'model' 컬럼이 존재합니다. 임시로 이름을 변경합니다.")
        tfidf_renamed_df.rename(columns={'model': 'tfidf_model_word'}, inplace=True)

    try:
        # 데이터프레임 길이를 맞추기 위해 인덱스 정렬 또는 리셋 후 concat
        # 원본 data에서 model 정보만 가져오고, tfidf 정보는 복사본 사용
        model_series = data['model'].reset_index(drop=True)
        tfidf_scores = tfidf_renamed_df.reset_index(drop=True)

        # 길이를 명시적으로 맞춤 (중요) - tfidf_df는 valid_answers_df 기준으로 만들어졌으므로 길이를 맞춰야 함
        if len(model_series) != len(tfidf_scores):
             # data도 valid_answers_df 기준으로 필터링된 버전을 사용해야 함
             valid_model_series = data.loc[valid_answers_df.index, 'model'].reset_index(drop=True)
             if len(valid_model_series) == len(tfidf_scores):
                 model_series = valid_model_series
             else:
                 raise ValueError(f"모델 정보와 TF-IDF 점수의 길이가 일치하지 않습니다. "
                                  f"data 길이: {len(data['model'])}, tfidf_df 길이: {len(tfidf_df)}, "
                                  f"필터링된 data 길이: {len(valid_model_series)}")

        # 모델 정보와 TF-IDF 점수 결합
        tfidf_with_model = pd.concat([model_series, tfidf_scores], axis=1)

        # 모델별로 각 단어의 평균 TF-IDF 점수 계산
        # 이제 'model' 컬럼은 하나만 존재하므로 그룹화 가능
        mean_tfidf_by_model = tfidf_with_model.groupby('model').mean().transpose()

        # 각 모델별 상위 N개 키워드 출력
        top_n = 15 # 상위 몇 개 키워드를 볼지 설정
        for model_name in mean_tfidf_by_model.columns:
            print(f"\n--- {model_name} (Top {top_n} Keywords) ---")
            # 점수가 0보다 큰 키워드만 대상으로 정렬 후 상위 N개 선택
            top_words = mean_tfidf_by_model[model_name][mean_tfidf_by_model[model_name] > 0].sort_values(ascending=False).head(top_n)
            pd.set_option('display.float_format', lambda x: '%.4f' % x) # 소수점 4자리 표시
            print(top_words)
            pd.reset_option('display.float_format') # 원래 포맷 복구

    except ValueError as e:
        print(f"오류 발생: {e}")
    except KeyError as e:
        print(f"오류 발생: '{e}' 컬럼을 찾을 수 없습니다. DataFrame 컬럼명을 확인하세요.")

else:
    print("오류: 'data' DataFrame에 'model' 컬럼이 없습니다.")
```

```python

# --- 4. 분석 2: 질문별 & 모델별 상위 TF-IDF 키워드 ---
print("\n\n--- 분석 2: 질문별 & 모델별 상위 TF-IDF 키워드 ---")
if all(col in data.columns for col in ['Question', 'model']):
    # 'data' DataFrame 사용
    unique_questions = data['Question'].unique()

    # 모든 질문 분석
    questions_to_analyze = unique_questions

    for question in questions_to_analyze:
        print(f"\n\n=== 질문: {question[:80]}... ===") # 질문이 길 경우 일부만 출력

        # 현재 질문에 해당하는 데이터의 인덱스 가져오기 ('data' 사용)
        question_indices = data[data['Question'] == question].index

        # 전체 TF-IDF 결과에서 해당 질문의 데이터만 선택
        # 인덱스가 유효한지 확인
        valid_indices = question_indices[question_indices.isin(tfidf_df.index)]
        if valid_indices.empty:
            print("(해당 질문에 대한 TF-IDF 데이터를 찾을 수 없습니다. 인덱스를 확인하세요.)")
            continue

        tfidf_subset_q = tfidf_df.loc[valid_indices]
        # 'data' DataFrame 사용
        model_subset_q = data.loc[valid_indices, 'model'] # 해당 질문의 모델 정보

        # 모델 정보와 TF-IDF 점수 결합 (인덱스 재설정으로 정렬)
        tfidf_with_model_q = pd.concat([model_subset_q.reset_index(drop=True), tfidf_subset_q.reset_index(drop=True)], axis=1)

        # 해당 질문 내에서 모델별 평균 TF-IDF 계산
        mean_tfidf_by_model_q = tfidf_with_model_q.groupby('model').mean().transpose()

        # 해당 질문에 대해 모델별 상위 N개 키워드 출력
        for model_name in mean_tfidf_by_model_q.columns:
            print(f"\n--- {model_name} (Top {top_n} Keywords for this Question) ---")
            # 점수가 0보다 큰 키워드만 대상으로 정렬 후 상위 N개 선택
            top_words_q = mean_tfidf_by_model_q[model_name][mean_tfidf_by_model_q[model_name] > 0].sort_values(ascending=False).head(top_n)
            if not top_words_q.empty:
                pd.set_option('display.float_format', lambda x: '%.4f' % x) # 소수점 4자리
                print(top_words_q)
                pd.reset_option('display.float_format')
            else:
                print("(No significant keywords found for this model in this question)")



print("\n\nTF-IDF 분석 완료.")
```

```python
# 필요한 라이브러리 설치 (아직 안 했다면)
# !pip install sentence-transformers scikit-learn

import pandas as pd
import numpy as np
import re
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from itertools import combinations # 모델 쌍 조합을 위해

# --- 이전 코드에서 combined_data 가 준비되었다고 가정 ---
# combined_data: model, Question, Attempt, Answer, Answer_clean 등 컬럼 포함

# --- SBERT 모델 로드 ---
print("SBERT 모델 로딩 중 ('all-MiniLM-L6-v2')...")
sbert_model = SentenceTransformer('all-MiniLM-L6-v2')
print("SBERT 모델 로딩 완료.")

# --- 답변 임베딩 생성 ---
# Answer_clean 컬럼의 모든 답변에 대해 임베딩 계산
print("답변 임베딩 생성 중...")
# NaN 값을 빈 문자열로 대체 (이미 처리되었을 수 있지만 확인차)
data['Answer_clean'] = data['Answer'].fillna('')
# 임베딩 계산 (시간이 다소 걸릴 수 있음)
embeddings = sbert_model.encode(data['Answer_clean'].tolist(), show_progress_bar=True)

# 계산된 임베딩을 DataFrame에 새로운 컬럼으로 추가 (리스트 형태로)
data['embedding'] = list(embeddings)
print("답변 임베딩 생성 완료.")

# --- 모델별/질문별 유사도 계산 ---

models = data['model'].unique()
questions = data['Question'].unique()

# 1. 모델 내부 유사도 (Within-Model Similarity)
print("\n--- 모델 내부 답변 유사도 계산 중 (동일 질문 내 답변 간 평균) ---")
within_model_similarity = {}

for model_name in models:
    model_similarities = []
    # 해당 모델의 데이터만 필터링
    model_data = data[data['model'] == model_name]

    for question in questions:
        # 해당 모델 & 해당 질문의 데이터 필터링
        question_data = model_data[model_data['Question'] == question]
        question_embeddings = np.array(question_data['embedding'].tolist())

        if question_embeddings.shape[0] > 1: # 답변이 2개 이상일 때만 유사도 계산 가능
            # 코사인 유사도 매트릭스 계산
            sim_matrix = cosine_similarity(question_embeddings)
            # 대각선(자기 자신과의 유사도=1) 제외하고 상단 삼각형(중복 제외) 값들만 추출
            upper_triangle_indices = np.triu_indices_from(sim_matrix, k=1)
            similarities = sim_matrix[upper_triangle_indices]

            if len(similarities) > 0:
                model_similarities.append(np.mean(similarities))
        # else: 답변이 1개 이하면 해당 질문의 유사도 계산 불가 (평균 계산 시 제외됨)

    # 모델의 모든 질문에 대한 평균 유사도 계산
    if model_similarities:
        within_model_similarity[model_name] = np.mean(model_similarities)
    else:
        within_model_similarity[model_name] = np.nan # 계산 가능한 유사도가 없는 경우

print("--- 모델 내부 평균 코사인 유사도 결과 ---")
for model_name, avg_sim in within_model_similarity.items():
    print(f"{model_name}: {avg_sim:.4f}")

## conda update jupyter
# 2. 모델 간 유사도 (Between-Model Similarity)
print("\n--- 모델 간 답변 유사도 계산 중 (동일 질문 내 답변 간 평균) ---")
between_model_similarity = pd.DataFrame(index=models, columns=models, dtype=float)

# 모든 모델 쌍에 대해 반복 (예: (gpt, gemini), (gpt, claude), ..., (deep_seek, human))
for model_a, model_b in combinations(models, 2):
    cross_model_similarities = []

    for question in questions:
        # 각 모델의 해당 질문에 대한 임베딩 추출
        embeddings_a = np.array(data[(data['model'] == model_a) & (data['Question'] == question)]['embedding'].tolist())
        embeddings_b = np.array(data[(data['model'] == model_b) & (data['Question'] == question)]['embedding'].tolist())

        if embeddings_a.shape[0] > 0 and embeddings_b.shape[0] > 0: # 두 모델 모두 해당 질문에 답변이 있을 때
            # 모델 A 답변들과 모델 B 답변들 간의 코사인 유사도 매트릭스 계산
            cross_sim_matrix = cosine_similarity(embeddings_a, embeddings_b)
            # 모든 쌍 간 유사도의 평균 계산
            cross_model_similarities.append(np.mean(cross_sim_matrix))
        # else: 한쪽이라도 답변 없으면 해당 질문 유사도 계산 불가

    # 두 모델 간의 모든 질문에 대한 평균 유사도 계산
    if cross_model_similarities:
        avg_cross_sim = np.mean(cross_model_similarities)
        between_model_similarity.loc[model_a, model_b] = avg_cross_sim
        between_model_similarity.loc[model_b, model_a] = avg_cross_sim # 대칭 행렬
    else:
        between_model_similarity.loc[model_a, model_b] = np.nan
        between_model_similarity.loc[model_b, model_a] = np.nan

# 대각선은 모델 내부 유사도로 채우기 (위에서 계산한 값 사용)
for model_name in models:
    between_model_similarity.loc[model_name, model_name] = within_model_similarity.get(model_name, np.nan)


print("\n--- 모델 간 평균 코사인 유사도 매트릭스 ---")
# 소수점 4자리까지 표시 설정
pd.set_option('display.float_format', lambda x: '%.4f' % x)
print(between_model_similarity)
# pd.reset_option('display.float_format') # 원래 포맷으로 복구 필요시
```
