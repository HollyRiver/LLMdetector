```python
import pandas as pd
from copy import deepcopy
```

```python
df_temp = pd.read_csv("temp_dist.csv")
dup = deepcopy(df_temp)
```

```python
for i in range(len(df_temp)) :
    for j in range(5) :
        if not pd.isna(dup.iloc[i, j]) :
            df_temp[i, j] = 0 ## 현재 자리 0으로
            df_temp.iloc[i, int(dup.iloc[i, j][1])] = float(dup.iloc[i, j][1:-1].split(", ")[-1])
```
